```python
import pandas as pd

# Cargamos el dataset limpio
df = pd.read_csv(
    r"C:\Users\santi\OneDrive\Escritorio\proyectos\datos_limpios.csv",
    low_memory=False
)

# 1. Tabla de producción mensual general
# Por qué: para el gráfico de evolución mes a mes
prod_mensual = df.groupby('mes')[['prod_pet', 'prod_gas', 'prod_agua']].sum().reset_index()
prod_mensual.columns = ['mes', 'petroleo_m3', 'gas_miles_m3', 'agua_m3']
prod_mensual.to_csv(
    r"C:\Users\santi\OneDrive\Escritorio\proyectos\powerbi_produccion_mensual.csv",
    index=False,
    sep=';',
    decimal=','
)

# 2. Tabla de producción por provincia
# Por qué: para el mapa o gráfico de barras por provincia
prod_provincia = df.groupby('provincia')[['prod_pet', 'prod_gas']].sum().reset_index()
prod_provincia.columns = ['provincia', 'petroleo_m3', 'gas_miles_m3']
prod_provincia.to_csv(
    r"C:\Users\santi\OneDrive\Escritorio\proyectos\powerbi_produccion_provincia.csv",
    index=False,
    sep=';',
    decimal=','
)

# 3. Tabla de producción por empresa
# Por qué: para el ranking de empresas
prod_empresa = df.groupby('empresa')[['prod_pet', 'prod_gas']].sum().reset_index()
prod_empresa.columns = ['empresa', 'petroleo_m3', 'gas_miles_m3']
prod_empresa = prod_empresa.sort_values('petroleo_m3', ascending=False)
prod_empresa.to_csv(
    r"C:\Users\santi\OneDrive\Escritorio\proyectos\powerbi_produccion_empresa.csv",
    index=False,
    sep=';',
    decimal=','
)

# 4. Tabla convencional vs no convencional
# Por qué: para mostrar el peso de Vaca Muerta
prod_recurso = df.groupby(['tipo_de_recurso', 'mes'])[['prod_pet', 'prod_gas']].sum().reset_index()
prod_recurso.columns = ['tipo_recurso', 'mes', 'petroleo_m3', 'gas_miles_m3']
prod_recurso.to_csv(
    r"C:\Users\santi\OneDrive\Escritorio\proyectos\powerbi_convencional_vs_no.csv",
    index=False,
    sep=';',
    decimal=','
)

# 5. Tabla de patrones de pozos
# Por qué: para el análisis de eficiencia y recomendaciones
patrones = pd.read_csv(
    r"C:\Users\santi\OneDrive\Escritorio\proyectos\patrones_pozos.csv"
)
patrones.to_csv(
    r"C:\Users\santi\OneDrive\Escritorio\proyectos\powerbi_patrones_pozos.csv",
    index=False,
    sep=';',
    decimal=','
)

print("Archivos para Power BI generados correctamente:")
print("- powerbi_produccion_mensual.csv")
print("- powerbi_produccion_provincia.csv")
print("- powerbi_produccion_empresa.csv")
print("- powerbi_convencional_vs_no.csv")
print("- powerbi_patrones_pozos.csv")
```
