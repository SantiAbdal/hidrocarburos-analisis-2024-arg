```python
import pandas as pd
from pathlib import Path
BASE = Path(__file__).parent if '__file__' in dir() else Path().resolve()

ruta_proceseed = BASE / "../../../data/processed/2025/datos_limpios.csv"
df = pd.read_csv(ruta_proceseed, low_memory=False)

print("=== PRODUCCIÓN TOTAL 2025 ===")
print(f"Petróleo (m3): {df['prod_pet'].sum():,.2f}")
print(f"Gas (miles m3): {df['prod_gas'].sum():,.2f}")
print(f"Agua (m3): {df['prod_agua'].sum():,.2f}")

print("\n=== PRODUCCIÓN MES A MES ===")
por_mes = df.groupby('mes')[['prod_pet', 'prod_gas']].sum()
print(por_mes)

print("\n=== PRODUCCIÓN POR PROVINCIA ===")
por_provincia = df.groupby('provincia')[['prod_pet', 'prod_gas']].sum()
por_provincia = por_provincia.sort_values('prod_pet', ascending=False)
print(por_provincia)

print("\n=== TOP 10 EMPRESAS - PETRÓLEO ===")
por_empresa = df.groupby('empresa')[['prod_pet', 'prod_gas']].sum()
por_empresa = por_empresa.sort_values('prod_pet', ascending=False)
print(por_empresa.head(10))

print("\n=== CONVENCIONAL VS NO CONVENCIONAL ===")
por_recurso = df.groupby('tipo_de_recurso')[['prod_pet', 'prod_gas']].sum()
por_recurso['pct_pet'] = (por_recurso['prod_pet'] / por_recurso['prod_pet'].sum() * 100).round(2)
por_recurso['pct_gas'] = (por_recurso['prod_gas'] / por_recurso['prod_gas'].sum() * 100).round(2)
print(por_recurso)

print("\n=== CONCENTRACIÓN DE MERCADO ===")
por_empresa_sorted = por_empresa.sort_values('prod_pet', ascending=False)
por_empresa_sorted['pct_acumulado'] = (
    por_empresa_sorted['prod_pet'].cumsum() / 
    por_empresa_sorted['prod_pet'].sum() * 100
)
empresas_80 = por_empresa_sorted[por_empresa_sorted['pct_acumulado'] <= 80]
print(f"Empresas que concentran el 80% de producción de petróleo: {len(empresas_80)}")
print(empresas_80[['prod_pet', 'pct_acumulado']])
```
