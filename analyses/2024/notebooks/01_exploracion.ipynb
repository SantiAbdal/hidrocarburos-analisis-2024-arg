```python
import pandas as pd
from pathlib import Path
BASE = Path(__file__).parent if '__file__' in dir() else Path().resolve()

# Cargamos el dataset
ruta_proceseed = BASE / "../../../data/raw/2024/produccin-de-pozos-de-gas-y-petrleo-2024.csv"
df = pd.read_csv(ruta_proceseed, low_memory=False)

# 1. Tamaño del dataset
print("=== TAMAÑO ===")
print(df.shape)

# 2. Tipos de datos de cada columna
print("\n=== TIPOS DE DATOS ===")
print(df.dtypes)

# 3. Cuántos nulos hay en cada columna
print("\n=== VALORES NULOS ===")
print(df.isnull().sum())

# 4. Valores únicos en columnas clave
print("\n=== VALORES ÚNICOS - TIPO DE RECURSO ===")
print(df['tipo_de_recurso'].unique())

print("\n=== VALORES ÚNICOS - TIPO DE POZO ===")
print(df['tipopozo'].unique())

print("\n=== VALORES ÚNICOS - TIPO ESTADO ===")
print(df['tipoestado'].unique())

print("\n=== VALORES ÚNICOS - TIPO EXTRACCION ===")
print(df['tipoextraccion'].unique())

print("\n=== VALORES ÚNICOS - PROVINCIA ===")
print(df['provincia'].unique())

# 5. Estadísticas básicas de las columnas numéricas
print("\n=== ESTADÍSTICAS BÁSICAS ===")
print(df[['prod_pet', 'prod_gas', 'prod_agua', 'iny_agua', 'iny_gas', 'iny_co2']].describe())
```
